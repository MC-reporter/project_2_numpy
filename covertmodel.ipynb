```python
import json
import numpy as np
import torch
from torch.utils.data import Dataset
from collections import OrderedDict
from json import JSONEncoder
%run network.ipynb

```

```python
class EncodeTensor(JSONEncoder,Dataset):
    def default(self, obj):
        if isinstance(obj, torch.Tensor):
            return obj.cpu().detach().numpy().tolist()
        return super(EncodeTensor, self).default(obj)
```

```python

def pthtojson(modelfile,jsonfile):
    obj = torch.load(modelfile)
    print(obj)
    with open(jsonfile, 'w') as json_file:
        json.dump(obj, json_file,cls=EncodeTensor)
    print("{} has been converted to {}".format(modelfile,jsonfile))
```

```python
def jsontonet(jsonfile,net):
    with open(jsonfile) as f:
        jsondata = f.read()
    dictkk =json.loads(jsondata)
    dd = OrderedDict()
    for k,v in dictkk.items():
        if isinstance(v, list):
            v = torch.tensor(np.asarray(v))
            dd[k] = v
    net.load_state_dict(dd)
    return net
```
